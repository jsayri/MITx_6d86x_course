A references for 3D plots with numpy is : https://matplotlib.org/3.1.1/gallery/mplot3d/scatter3d.html

```python
import numpy as np
import matplotlib.pyplot as plt
```

We define a bunch of negative labels around a radius of 1 to 1.5 and positive labels within 3 and 4 as radius

```python
# generation of 2-dimensional feature vector
num_neg = 100
rn_min = 1
rn_max = 2
tn = np.random.random(num_neg) * 2 * np.pi
rn = rn_min + (rn_max-rn_min) * np.random.random(num_neg)
xn = rn * [np.cos(tn), np.sin(tn)]
yn = -1*np.ones(num_neg)

num_pos = 100
rp_min = 3
rp_max = 4
tp = np.random.random(num_pos) * 2 * np.pi
rp = rp_min + (rp_max-rp_min) * np.random.random(num_pos)
xp = rp * [np.cos(tp), np.sin(tp)]
yp = np.ones(num_pos)
```

```python
# 2-dimensional feature vectors
x = np.concatenate((xn.transpose(), xp.transpose()))
y = np.concatenate((yn, yp))
```

```python
# creation of the 3-dimension axe
fun_3d = lambda x1, x2: x1**2 + x2**2
```

```python
fig = plt.subplots(ncols=2, figsize=(15,7))

# 2-d representation
ax1 = plt.subplot(121)
ax1.plot(xn[0,:], xn[1,:], 'bx')
ax1.plot(xp[0,:], xp[1,:], 'go')
ax1.set_xlabel('Dimension 1')
ax1.set_ylabel('Dimension 2')

# 3-d representation
ax2 = plt.subplot(122, projection='3d')
ax2.scatter(xn[0,:], xn[1,:], fun_3d(xn[0,:], xn[1,:]), marker = 'x', color = 'blue')
ax2.scatter(xp[0,:], xp[1,:], fun_3d(xp[0,:], xp[1,:]), marker = 'o', color = 'green')
ax2.set_xlabel('Dimension 1')
ax2.set_ylabel('Dimension 2')
ax2.set_zlabel('Dimension 3')

ax2.view_init(20, 45)
```

So a $f(x_1,x_2)$ was selected to build the 3D representation of the points in 2D.

A solution in 2D that will define a boundary in $z=4$ is $f(x_1,x_2) = 0 = x1^2 + x2^2 - 4$

```python
# define plot
fig = plt.figure(figsize=(7,9))
ax = fig.add_subplot(111, projection='3d')

# check function f(x1,x2) for a boundary at z=4
f_boundary = lambda x: x - 4

# display results based on boundary function f(x1,x2) = 0
isZero = 10**-10
for xr in x:
    x3 = fun_3d(xr[0], xr[1])
    if f_boundary(x3) <= isZero:
        ax.scatter(xr[0], xr[1], x3, marker = 'x', color = 'blue')
    else:
        ax.scatter(xr[0], xr[1], x3, marker = 'o', color = 'green')
        
ax.view_init(20, 45)
ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.set_zlabel('Dimension 3')
```
